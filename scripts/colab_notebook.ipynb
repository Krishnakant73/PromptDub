## Step 5: Start ngrok Tunnel

```
from pyngrok import ngrok
import time

# Free ngrok account gives 1 tunnel
public_url = ngrok.connect(8000, "http")
print(f"\n{'='*60}")
print(f"PUBLIC URL: {public_url}")
print(f"{'='*60}")
print(f"\nCopy this URL into your Chrome extension popup:")
print(f"  wss://{public_url.netloc}/ws/translate")
print(f"\nExample: wss://abc123.ngrok-free.app/ws/translate")
```

## Connect Chrome Extension

1. Open the PromptDub extension popup
2. Paste the WebSocket URL (e.g., `wss://abc123.ngrok-free.app/ws/translate`)
3. Select target language
4. Open a YouTube/Twitch livestream
5. Click **Start Translating**
